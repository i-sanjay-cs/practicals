```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Wine.csv')
```

```python
df
```

```python
x = df.drop('Customer_Segment',axis=1)
y = df['Customer_Segment']
```

```python
sc = StandardScaler()
standard = sc.fit_transform(x)
```

```python
pca = PCA()
pca.fit_transform(standard)
```

```python
explained_variance = pca.explained_variance_ratio_
plt.plot(range(1,len(explained_variance)+1),explained_variance.cumsum(),marker='o',linestyle='--')
```

```python
pca = PCA(n_components=0.95)
x_pca = pca.fit_transform(standard)
```

```python
plt.scatter(x_pca[y==1][:,0],x_pca[y==1][:,1], c = 'r',label = 'Red Wine')
plt.scatter(x_pca[y==2][:,0],x_pca[y==2][:,1], c = 'b',label = 'Blue Wine')
plt.xlabel('Principal component one')
plt.ylabel('Principal component two')
plt.legend()
plt.show()
```
