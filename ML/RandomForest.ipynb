```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Step 1: Load and preprocess the dataset
# Load the dataset into a DataFrame
df = pd.read_csv('car_evaluation.csv')

# Encode categorical variables
label_encoders = {}
categorical_columns = ['buying_price', 'maintenance_cost', 'lug_boot']
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Encode the 'safety' column (target variable)
le_safety = LabelEncoder()
df['safety'] = le_safety.fit_transform(df['safety'])
label_encoders['safety'] = le_safety


```

```python
# Separate the target variable
X = df.iloc[:, [0, 1]]  # Features (assuming 'buying_price', 'maintenance_cost', and 'lug_boot' are the first three columns)
y = df['safety']  # Target

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Random Forest Model
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Step 4: Train the Model
rf_classifier.fit(X_train, y_train)

# Step 5: Model Evaluation
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

```

```python

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```
