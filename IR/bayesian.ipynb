```python
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
df = pd.read_csv('heart2.csv')
```

```python
df
```

```python
model = BayesianNetwork([('age','target'),
                         ('sex','target'),
                         ('cp','target'),
                         ('trestbps','target'),
                         ('chol','target'),
                         ('fbs','target'),
                         ('restecg','target'),
                         ('thalach','target'),
                         ('exang','target'),
                         ('oldpeak','target'),
                         ('slope','target'),
                         ('ca','target'),
                         ('thal','target')
                        ])
```

```python
model.fit(data=df,estimator=MaximumLikelihoodEstimator)
```

```python
evidence = {
    'age':55,
    'sex':1,
    'cp':0,
    'trestbps':140,
    'chol':212,
    'fbs':0,
    'restecg':1,
    'thalach':122,
    'exang':0,
    'oldpeak':1.0,
    'slope':2,
    'ca':2,
    'thal':3
}
```

```python
inference = VariableElimination(model)
```

```python
query = inference.query(variables=['target'],evidence=evidence)
print(query)
```

```python
# Diagnose the patient based on the probability
if query.values[1] > query.values[0]:
    print("The patient is likely to have Heart Disease.")
else:
    print("The patient is likely not to have Heart Disease.")
```
